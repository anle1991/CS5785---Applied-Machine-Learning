## Import data

```python
import pandas as pd
data_set = pd.read_csv('iris.data', header=None, 
                       names=['Sepal Length','Sepal Width', 
                              'Petal Length', 'Petal Width', 'Class'])
print(data_set[:5])
```

```python
print("Shape of dataset: {}".format(data_set.shape))
```

```python
data_set['Class'].value_counts()
```

## Convert to narray and split

```python
column_names = ["Sepal Length","Sepal Width","Petal Length","Petal Width"];
data = data_set.as_matrix(column_names)
target = data_set.as_matrix(["Class"])

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state = 0)
```

## Plot

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

iris_dataset = pd.DataFrame(X_train, columns=column_names)

colors = []
for e in y_train:
    if e=='Iris-setosa':
        colors.append('r')
    elif e=='Iris-versicolor':
        colors.append('b')
    elif e=='Iris-virginica':
        colors.append('g')
    
pd.plotting.scatter_matrix(iris_dataset, figsize=(15,15), 
                           marker='o', hist_kwds={'bins': 20}, s=60, 
                           alpha=.8, c=colors)
```
